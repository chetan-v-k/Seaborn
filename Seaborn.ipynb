Streamlining the Categorical Visualization with Countplots
---------------------------------------------------


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins.info()
```

```python
penguins['species'].value_counts()
```

```python
#Create a counplot to show species distribution
sns.countplot(data=penguins, x='species', hue = 'sex')
#x= - column that you want to show your x-axis
plt.title('Count of Penguin Species')
plt.xlabel('Species')
plt.ylabel('count')
plt.show()   # guarantees the proper rendering of the visualizations
```

Effortlessly Enhancing the Histograms(histplot)  on Seaborn

```python
sns.histplot(data=penguins, x='body_mass_g', bins = 30, kde=True)   #KDE- Kernel density Estimation
plt.title('Penguin body mass distribution')
plt.xlabel('body_mass(g)')
plt.ylabel('frequency')
plt.show()
```

Unlocking the power of Seaborn Pairplots

```python
sns.pairplot(data=penguins, vars=['bill_length_mm','flipper_length_mm','body_mass_g'], hue='species')   #pairplot with selected variables
#vars - selected variables are bill_length_mm & flipper_length_mm & body_mass_g
plt.show()

#NOTE: sns.pairplot() expects only numerical or datetime variables in the vars parameter for plotting
# If both bill_length_mm and flipper_length_mm are numerical:
# It will show a scatter plot between them.
# It will also show histograms for individual distributions on the diagonal.

#After adding the hue the histograms are converted into the Kernel Density Estimation Curves(KDE)
#by removing the vars parameter. This change will help you visualize all variables in the dataset
```

Analyzing the Data Distribution with Seaborn boxplots

```python
#boxplots are excellent for outliers detection

#Create a boxplot to visualize the distribution of flipper lengths for each species
sns.boxplot(data=penguins,x= 'species', y='flipper_length_mm',hue= 'sex') # species along x-axis and flipper_lenght_mm along y-axis
plt.title('Boxplot of flipper Lengths of each penguins species with sex as hue')
plt.ylabel('Flipper Length (mm)')
plt.xlabel('Species')
plt.show()
```

Exploring the Multidimensional Data with Seaborn FacetGrid

For Multi-Plot visualization - Comparative Analysis - Navigate Complex Datasets

```python
#FacetGrid with each species of penguins in different columns
g=sns.FacetGrid(data=penguins, col='species', row='sex')

#map histograms with kde to each species columns
g.map(sns.histplot,'flipper_length_mm',kde=True)
plt.show()
```
